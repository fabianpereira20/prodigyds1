```python
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
df = pd.read_csv("population.csv")
df.head()
```

```python
df = df.drop("Country Code",axis=1)
df = df.drop("Series Name",axis=1)
df.head()
```

```python
Countries_plot = ['China','India','United States','Indonesia','Pakistan','Nigeria','Brazil']
df_plot = df[df['Country Name'].isin(Countries_plot)]
year = '2022'
plt.figure(figsize=(10,6))
plt.bar(df_plot['Country Name'],df_plot[year])
plt.xlabel('country Name')
plt.ylabel(f'population of {year}')
plt.title(f'population of selected countries in {year}')
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
def add_value_labels(ax, spacing=10):

    # For each bar: Place a label
    for rect in ax.patches:


        x = rect.get_x() + rect.get_width() / 2
        y = rect.get_height()-3


        va = 'bottom' if y >= 0 else 'top'


        label = "{}".format(y)


        y_shift = spacing * (1 if y >= 0 else -1)


        ax.annotate(label, (x, y), xytext=(0, y_shift),textcoords="offset points", ha='center', va=va)
```

```python
male_population_data = df[df['Series Code'] == "SP.POP.TOTL.MA.IN"]
male_population_sorted = male_population_data.sort_values(by='2022', ascending=False)
male_top_8 = male_population_sorted.head(8)
print("TOP 8 countries with male populatiion")
print(male_top_8)
```

```python

female_population_data = df[df["Series Code"] == "SP.POP.TOTL.FE.IN"]
female_population_sorted = female_population_data.sort_values(by="2022", ascending=False)


female_top_8 = female_population_sorted.head(8)

print("Top 8 countries with female population")
print(female_top_8)
```

```python
plt.figure(figsize=(20,16))
plt.rcParams['axes.facecolor'] = 'white'
plt.subplot(2,2,1)
plot =sns.barplot(x="2022",y="Country Name",data = male_top_8,orient = "h")
plt.title("Top 8 countries with male population")
plt.xlabel("male population")
plt.ylabel("Country Name")
add_value_labels(plot)

plt.subplot(2,2,2)
plot=sns.barplot(x="2022",y="Country Name",data = female_top_8,orient = "h")
plt.title("Top 8 countries with female population")
plt.xlabel("female population")
plt.ylabel("Country Name")
add_value_labels(plot)

plt.tight_layout()
plt.show()
```
